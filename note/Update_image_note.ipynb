## Note on Updating Pod Image 
When update docker image for pods, the cluster does not have access to new image in docker
=> need to push the new image to online Docker repository or push it  

```python
#  rebuild image
docker build -t ghcr.io/giangnguyen2007/myspark:v3 .

# push new image to repo
docker push ghcr.io/giangnguyen2007/myspark:v3

# or load the new image into the cluster nodes
minikube image load myrepo/spark-numpy:1.0
```
